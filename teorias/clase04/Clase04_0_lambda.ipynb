# Repaso 4

¿Qué diferencias hay entre las siguientes funciones?

```python
def my_args(*args):
    ....
def my_args1(**kwargs):
    ....
def my_args2(a, b, /, *, exacto=False):      

```

- **args** es una **tupla** que representa a los  parámetros pasados.
- **kwargs** es una **diccionario** que representa a los  parámetros pasados.
- Se puede organizar los parámetros
  - **Parámetros posicionales exclusivos**  (antes de **/**): los argumentos para estos parámetros deben pasarse únicamente por posición. No se puede usar el nombre del parámetro en la llamada.
  - **Parámetros con nombre exclusivos** (después de **\***): los argumentos para estos parámetros deben pasarse obligatoriamente usando su nombre (keyword arguments). No se pueden pasar de forma posicional.

# DESAFÍO 1
> Queremos implementar una función que dada una cadena de texto, retorne las palabras que contiene en <span style="color:#cd7815">orden alfabético</span>.

````python
# Una posible solución
def sort1(sentence):
    """ Example using method sort"""

    words = sentence.split()
    words.sort(key=str.lower)
    #words.sort()
    return words
````

````python
print(sort1("Hoy puede ser un gran día. "))
````

¿De qué tipo es el valor enviado como parámetro a **key**?

# Otra forma

````python
def sort2(sentence):
    """ Example using function sorted"""

    words = sentence.split()
    return sorted(words, key=str.lower)
````

````python
print(sort2("Hoy puede ser un gran día. "))
````

# La función map

````python
def double(x):
    return 2*x
````

````python
numbers = [1, 2, 3, 4, 5, 6, 7]
doubles = map(double, numbers)
for elem in doubles:
    print(elem, end="  ")
````

# La función filter

````python
def is_even(x):
    return x%2 == 0
````

````python
numbers  = [1, 2, 3, 4, 5, 6, 7]
even_numbers = filter(is_even, numbers)

for elem in even_numbers:
    print(elem, end="  ")
````

# map y filter con  lambda

````python
numbers  = [1, 2, 3, 4, 5, 6, 7]

double_numbers = map(lambda x: 2*x, numbers)
even_numbers = filter(lambda x: x%2 == 0, numbers)
````

````python
for elem in double_numbers:
    print(elem, end="  ")
````

# La función reduce  

Se aplica una función con los dos primeros elementos de la secuencia iterable, generando un resultado parcial. Luego, aplica nuevamente la función con este resultado parcial y el tercer elemento, repitiendo lo mismo para todos los elementos. 

````python
from functools import reduce

numbers  = [1, 2, 3, 4, 5, 6, 7]

reduce(lambda a, b: a + b, numbers)
min_value = reduce(lambda a, b: a if a < b else b, numbers)
````

- ¿Cuántos elementos se obtienen luego de aplicar reduce?
- ¿Qué sucede se la secuencia está vacía?

````python
reduce(lambda a, b: a + b, numbers)
````

> Tenemos una lista de temperaturas en grados Farenheit y queremos generar otra lista con las temperaturas en grados Celcius.

Recordemos que la fórmula es: C = (F -32)/1.8


````python
# Posible solución SIN lambda
fahrenheit = [32, 50, 68, 212]
celsius = [(f - 32)/1.8 for f in fahrenheit]
celsius
````

````python
# Posible solución CON lambda
fahrenheit = [32, 50, 68, 212]
celsius = list(map(lambda f: (f - 32)/1.8, fahrenheit))
celsius
````

nombre: Sofía apellido: Martin, titulo: Prof

Listado:
