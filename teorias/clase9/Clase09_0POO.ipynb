# ¿@property?

> <span style="color:#cd7815">@property permite crear **propiedades** de una clase.</span>

- Se aplica la función **property()** cuya forma general es: 


```python
	property(fget=None, fset=None, fdel=None, doc=None)
```

¿En qué casos podemos usar propiedades?

# Ejemplo

````python
class Demo:
    def __init__(self):
        self._x = 0

    def get_x(self):
        print('Estoy en get')
        return self._x

    def set_x(self, value):
        print('Estoy en set')
        self._x = value
    
    x = property(get_x, set_x)
````

````python
obj = Demo()
obj.x = 10
print(obj.x)
````

# O en su forma equivalente ... 

````python
class Demo:
    def __init__(self):
        self._x = 0
    @property
    def x(self):
        print('Estoy en get')
        return self._x

    @x.setter
    def x(self, value):
        print('Estoy en set')
        self._x = value
````

````python
obj = Demo()
print(obj.x)
````

- @property es un <span style="color:#cd7815">**decorador**</span>.

## Métodos de clase
¿A qué creen que hacen referencia?

> <span style="color:#cd7815">Corresponden a los mensajes que se envían a la **clase**, no a las instancias de la misma.</span>

- Se utiliza el<span style="color:#cd7815"> decorador **@classmethod**</span>.
- Se usa **cls** en vez de **self**. ¿A qué hace referencia este argumento? 

````python
class Band():
    """ Define la entidad que representa a una banda   ..   """
    all_genres = set()
    
    @classmethod
    def clean_genres(cls, confirm=False):
        if confirm:
            cls.all_genres = set()
            
    def __init__(self, name, genres="rock"):
        self.name = name
        self.genres = genres
        self._members = []
        Band.all_genres.add(genres)
  
    def add_member(self, new_member):
        self._members.append(new_member)
````

## Métodos estáticos
> <span style="color:#cd7815">Son funciones definidas dentro de una clase que **no acceden ni modifican el estado de la instancia (self) ni de la clase (cls)**.</span> 

- Se utiliza el<span style="color:#cd7815"> decorador **@staticmethod**</span>.
- NO se usa ni **cls** ni **self** ya que no reciben argumentos automáticos y se utilizan para funciones "útiles" que lógicamente pertenecen a la clase pero no corresponden al comportamiento de los objetos. 

## Veamos esto en un ejemplo

````python
class Band():
    """ Define la entidad que representa a una banda """
            
    def __init__(self, name, genres="rock"):
        # Usamos el método estático para normalizar el nombre al nacer
        self.name = self.format_name(name)
        self.genres = genres
        self._members = []

    def add_member(self, new_member):
        self._members.append(new_member)

    # --- MÉTODO ESTÁTICO ---
    @staticmethod
    def format_name(name):
        """
        Quita espacios extra y pone mayúsculas iniciales en el nombre de la banda.
        IMPORTANTE: notar que NO usa 'self' ni 'cls'.
        """
        return name.strip().title()
````

````python
# 1. Uso sin crear una instancia (Utilidad pura)
real_name = "  the rolling stones  "
clean_name = Band.format_name(real_name)
print(f"Nombre procesado: '{clean_name}'") 
````

````python
# 2. Uso dentro de la clase
new_band = Band("    rage Against the machine        ", "rock")
print(f"Banda creada: {new_band.name}") 
````

# Hablemos de herencia

- Es uno de los conceptos más importantes de la POO.
- La herencia permite que una clase pueda *heredar* los atributos y métodos de otra clase, que se **agregan** a los propios. 
- Este concepto permite sumar, es decir <span style="color:#cd7815">**extender** </span>una clase.
- La clase que hereda se denomina **clase derivada** y la clase de la cual se deriva se denomina **clase base**.

# El Name Mangling

> Es un mecanismo para evitar colisiones de nombres en la herencia.

## ¿Qué era esto?
Cuando Python ve un atributo llamado **\_\_atributo** dentro de una clase, lo renombra internamente siguiendo esta fórmula:

\_NombreDeLaClase\_\_nombre_del_atributo

- Recordemos que esto no implica que la variable es privada. 

## ¿Cuándo se aplica?
- Nombres que comienzan con dos "\_"  y terminen con ningún o un único guión bajo.
- ¿Por qué esto no aplica a \_\_init\_\_, \_\_str\_\_ etc.?


# Habíamos visto este ejemplo

````python
class SecretCode:
    """¿¿¿Textos con clave??? """

    def __init__(self, plain_text, secret_key):
        self.__plain_text = plain_text
        self.__secret_key = secret_key

    def decrypt(self, secret_key):
        """Solo se muestra el texto si secret_key es correcta"""
        
        if secret_key == self.__secret_key:
            return self.__plain_text
        else:
            return ''
````

## ¿Para qué se usa entonces?

- La razón principal no es esconder datos del usuario, sino proteger a las subclases.

- Observemos el siguiente ejemplo:

````python
class Vehículo:
    def __init__(self, cantidad_puertas, cantidad_ruedas, motor='Diesel'):
        self.cant_puertas = cantidad_puertas
        self.cant_ruedas = cantidad_ruedas
        self.motor = motor
        # Identificador interno de fábrica (Mangling)
        self.__id_interno = "FAB-12345"
        
    def __str__(self):
        return f'Motor: {self.motor}, puertas: {self.cant_puertas}, ruedas: {self.cant_ruedas}'
        
class Transporte():
    def __init__(self, max_pasajeros, tipo='Público'):
        self.tipo = tipo
        self.cant_max_pasajero = max_pasajeros
        # Identificador interno administrativo (Mangling)
        self.__id_interno = "ADM-99999"
        
    def __str__(self):
        return f'Tipo de transporte: {self.tipo}'
````

````python
class Automóvil(Vehículo, Transporte):
    def __init__(self, marca, modelo, patente, tipo='Particular', pasajeros=0, puertas=4, ruedas=4, motor='Nafta'):
        Vehículo.__init__(self, puertas, ruedas, motor)
        Transporte.__init__(self, pasajeros, tipo)
        self.marca = marca
        self.modelo = modelo
        self._patente = patente
````

````python
auto = Automóvil("Toyota", "Corolla", "ABC-123")

# Da error al intentar acceder directamente 
#print(auto.__id_interno)

# Se puede acceder a ambos atributos "mangled"
print("Atributos del Automóvil:")
print(f"ID de Fábrica: {auto._Vehículo__id_interno}")
print(f"ID Administrativo: {auto._Transporte__id_interno}")
````

# Un último ejemplo: esto aplica también a métodos

````python
class Transporte():
    def __init__(self, max_pasajeros, tipo='Público'):
        self.tipo = tipo
        self.cant_max_pasajero = max_pasajeros
        self.__id_interno = "ADM-99999"
        
    def __str__(self):
        return f'Tipo de transporte: {self.tipo}'

    def _validar_capacidad(self, actuales):
        """Lógica para saber si suben más pasajeros."""
        return actuales <= self.cant_max_pasajero

    def __actualizar_seguro_transporte(self):
        """Trámite interno de la administración."""
        print(f"Actualizando póliza para transporte tipo {self.tipo}...")
````

¿Qué diferencias hay entre los métodos **\_validar\_capacidad** y **\_\_actualizar\_seguro\_transporte**?

Al igual que con las variables de instancia:
- **metodo():** accesible desde cualquier lugar (dentro y fuera de la clase)
- **\_metodo():** indica uso interno. Por convención se asume privado. 
- **\_\_metodo():** Name Mangling: Python renombra el método a _Clase__metodo para evitar colisiones en herencia.

````python
class Musician:
    def __init__(self, name, instrument=None, band=None):
        self.name = name
        self.has_a_band = band!=None
        self._band = band
        self.instrument = instrument
        
    def __add__(self, other):
        return (f"Nuevo dúo: {self.name} y {other.name}")    
    
    @property
    def band(self):
        if self.has_a_band:
            return self._band
        else:
            return None
        
    @band.setter
    def band(self, new_band):
        self._band = new_band
        self.has_a_band= self._band!=None
````

# Creamos iteradores

> <span style="color:#cd7815">Un **iterador** es un objeto que permite recorrer **uno a uno** los elementos de una estructura de datos para poder operar con ellos.</span>

- Un objeto iterable tiene que implementar un <span style="color:#cd7815">método **\_\_next\_\_**</span> que debe devolver los elementos, **de a uno por vez**, comenzando por el primero. 
- Y al llegar al final de la estructura, debe levantar la <span style="color:#cd7815">excepción **StopIteration**</span>.

- Todas implementan un método especial denominado <span style="color:#cd7815">**\_\_iter\_\_**</span>. 
    -  **\_\_iter\_\_** devuelve un iterador capaz de recorrer la secuencia.

# Los siguientes códigos son equivalentes:

````python
lista = ['uno', 'dos', 'tres']
for palabra in lista:
    print(palabra)
````

````python
iterador = iter(lista)
while True:
    try:
        palabra = next(iterador) # o iterador.__next__()
    except StopIteration:
        break
    print(palabra)
````

- La función <span style="color:#cd7815">**iter**</span> retorna un objeto iterador.

¿Son equivalentes?

````python
for elem in lista:
    print (elem)
````

````python
for elem in iterador:
    print (elem)
````

# Veamos este ejemplo: 

````python
class CadenaInvertida:
    def __init__(self, cadena):
        self._cadena = cadena
        self._posicion = len(cadena)

    def __iter__(self):
        return(self) 

    def __next__(self):
        if self._posicion == 0:
            raise(StopIteration)
        self._posicion = self._posicion - 1
        return(self._cadena[self._posicion])
````

````python
cadena_invertida = CadenaInvertida('La clase próxima empezamos con pandas')

for caracter in cadena_invertida:
     print(caracter, end=' ')
````

>Implementar la clase  **CadenaCodificada**, que dada una cadena de caracteres, me permita trabajar con la misma en forma codificada, según la codifcación Cesar.  Podemos recorrer con un **for** un objeto de clase **CadenaCodificada**, los cual permite acceder uno a uno a los caracteres codificados de la misma.
>
>Ejemplo: si la cadena orginal es "Hola que tal", la cadena codificada será "Ipmb!rvf!ubm"

**Nota:** implementar este objeto como un objeto iterable. 

````python
#Una posible solución que no vamos a mostrar pero si a ejecutar como para que vean a qué nos referimos
from functools import reduce

class CadenaCodificada:
    def __init__(self, cadena):
        self._cadena = cadena
        self._posicion = 0

    def __iter__(self):
        return(self) 

    def __next__(self):
        if self._posicion == len(self._cadena):
            raise(StopIteration)
        car = self._cadena[self._posicion]
        self._posicion = self._posicion + 1 
        return(chr(ord(car) + 1))
    
    def __str__(self):
        lista = map(lambda c : chr(ord(c) + 1), self._cadena)
        return reduce(lambda c1, c2: c1 + c2, lista)
````

## Ejemplo de uso de una CadenaCodificada

````python
mi_cadena = CadenaCodificada("Hola que tal")

for caracter in mi_cadena:
     print(caracter, end=" ")
````
