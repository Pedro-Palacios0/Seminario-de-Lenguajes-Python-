## Enunciado a evaluar 


**EJERCICIO 1.** Escribir una función que retorne los 5 usuarios que más veces accedieron al material de clases  durante la mañana, y los horarios en los que accedieron. Esta función recibe un parámetro opcional que indica si los nombres de usuarios se muestran en mayúsculas **("M")**, minúsculas **("m")** o como aparecen en el dataset **(None)**. Por defecto la función retorna los nombres tal como aparecen en el dataset. 

**NOTA:** la información sobre el material de clases se obtiene verificando el texto "Página: Material de las clases" en la columna "Contexto del evento" del dataset.

Consideramos que el horario de mañana es entre las 06:00 y las 12:00.

**EJERCICIO 2.** Utilizando la función anterior, escribir un programa que muestre dos listados correspondientes a la actividad  de los 5 usuarios que más veces accedieron al material de clases durante la mañana: uno en el cual  se muestren los nombres de usuario en minúsculas y otro tal cual como aparecen en el dataset.

### Solución a analizar

**EJERCICIO 1**

```python
import csv
from pathlib import Path 
from collections import Counter 

def mas_accesos(csv_reader, formato=None):
    accesos = filter(lambda x: x["Contexto del evento"] == "Pagina: Material de las clases" and "06:00:00"
                                <= x["Hora"] <= "12:00:00", csv_reader)
    
    alumnos = map(lambda x: (x["alumnos"].upper() if formato == "M" else
                             x["alumnos"].lower() if formato == "m" else
                             x["alumnos"]), accesos)
    
    return Counter(alumnos).most_common(5)
```

**EJERCICIO 2**

```python
file_route = Path('archivos' / "logs_catedra.csv")
file = open(file_route,"r")
csv_reader = csv.DictReader(file, delimiter = ',')

top5_alumnos = mas_accesos(csv_reader)

print(f"{'Alumno':<30}{'Accesos'}")
print('-' * 40)

for alumno,accesos in top5_alumnos:
    print(f"{alumno:<30}{accesos}")

print("Alumnos con mas accesos en minúsuculas: ")
print(f"{'Alumno':<30}{'Accesos'}")
print('-' *40)

for alumno,accesos in top5_alumnos:
    print(f"{alumno.lower():<30}{accesos}")

file.close()
    
```

```python
#Solución dada por la IA

import pandas as pd

# Ejercicio 1
def top5_material_manana(ruta, formato=None):
    df = pd.read_csv(ruta)
    df["hora"] = pd.to_datetime(df["hora"])
    
    material = df[df["Contexto del evento"].str.contains("Página: Material de las clases", na=False)]
    manana = material[(material["hora"].dt.hour >= 6) & (material["hora"].dt.hour < 12)]
    
    conteo = manana.groupby("alumnos").size()
    top5 = conteo.nlargest(5).index.tolist()
    
    resultado = {}
    for alumno in top5:
        horarios = manana[manana["alumnos"] == alumno]["hora"].tolist()
        if formato == "M":
            nombre = alumno.upper()
        elif formato == "m":
            nombre = alumno.lower()
        else:
            nombre = alumno
        resultado[nombre] = horarios
    
    return resultado


# Ejercicio 2
ruta = "ruta/del/archivo.csv"

resultado_minus = top5_material_manana(ruta, formato="m")
print("=== Nombres en minúsculas ===")
for alumno, horarios in resultado_minus.items():
    print(f"\n{alumno}:")
    for h in horarios:
        print(f"  {h}")

resultado_original = top5_material_manana(ruta, formato=None)
print("\n=== Nombres como aparecen en el dataset ===")
for alumno, horarios in resultado_original.items():
    print(f"\n{alumno}:")
    for h in horarios:
        print(f"  {h}")
```
